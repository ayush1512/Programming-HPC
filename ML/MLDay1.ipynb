```python
# Sets
```

```python
set1 = {1,2,3,4,5}
type(set1)
```

```python
#Add
set1.add(6)
set1
```

```python
set1.remove(3)
set1
```

```python
## Set Operations
### Union
set1 = {1,2,3,4,5}
set2 = {4,5,6,7,8}
# set3 = set1.union(set2)
set3 = set1 | set2
set3
```

```python
### Intersect
set4 = set1.intersection(set2)
set4 = set1 & set2
set4
```

```python
### Difference
set5 = set1 - set2
set5
```

```python
set6 = set2 - set1
set6
```

```python
### Symmetric Difference
set7 = set1 ^ set2
set7
```
