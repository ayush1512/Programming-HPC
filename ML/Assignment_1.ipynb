# 9. Grouping Data Group a DataFrame of sales by region and find total sales in each region.

```python
reg = ['N','S','W','E']
regn = np.random.choice(reg,size=(10))
sales = np.random.randint(100000,1000000,size=(10))
salesReg = pd.DataFrame({'Region':regn,'Sales':sales})
salesReg

```

```python
salesByReg = salesReg.groupby('Region')['Sales'].sum()
salesByReg
```

# 10. Data Merge Merge two DataFrames:
- students (id, name)
- scores (id, marks) Display merged data showing student name and marks.

```python
id = np.arange(1,11)
name = [chr(i) for i in range(65,75)]
marks = np.random.randint(33,99,size=(10))
students = pd.DataFrame({'id':id,'name':name})
scores = pd.DataFrame({'id':id,'marks':marks})
```

```python
merged = pd.merge(left=students,right=scores,on='id')
merged
```

# 11. Basic Line Plot Create a line plot for monthly sales data. Label axes and title the plot “Monthly Sales Trend”.

```python
import calendar
import matplotlib.pyplot as plt
months = list(calendar.month_abbr)
months.remove('')
sales = np.random.randint(10000,1000000,size=(12))

```

```python
plt.plot(months,sales)
plt.figaspect(10)
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title("Monthly Sales Trend")
plt.show()
```

# 12. Bar and Pie Plot Plot a bar chart showing the number of students per department (e.g., CS, IT, ECE). Also, represent the same data using a pie chart.

```python
br = ['CS','IT','ECE','CE','ME']
branch = np.random.choice(br,size=(20))
branch
```
