```python
op = int(input("Enter option (1-4): "))

match op:
    case 1:
        print("Option 1: Add")
        arr.append(int(input("Enter value: ")))
    case 2:
        print("Option 2: Remove")
        arr.pop() if arr else print("Empty!")
    case 3:
        print("Option 3: Display")
        print(arr)
    case 4:
        print("Exit")
    case _:
        print("Invalid option")
```
