# 2. Demonstrate Central Limit Theorem using the following dataset: 5 Marks
'''
values = np.array([10, 20, 30, 40, 50, 60, 70])
freq = np.array([2000, 2500, 3000, 1500, 800, 150, 50])
population = np.repeat(values, freq)
'''

```python
values = np.array([10, 20, 30, 40, 50, 60, 70])
freq = np.array([2000, 2500, 3000, 1500, 800, 150, 50])
population = np.repeat(values, freq)

s_mean = []
s_size = 50
samples = 1000

for _ in range(samples):
    sample = np.random.choice(population, s_size, replace=False)
    s_mean.append(np.mean(sample))

plt.figure()
sns.histplot(s_mean, kde=True, color='blue', bins=30)
plt.title("Distribution")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
```

# 3. A machine in a factory produces electronic components, and 3 Marks
'''
each component has a 15% chance of being defective.
If the machine produces 12 components in a day, find the probability that:
a. Exactly 3 components are defective.
b. At most 3 components are defective.
c. At least 3 components are defective.
'''

```python
p_def = 0.15
component = 12
p_exactly_3 = binom.pmf(3, component, p_def)
p_at_most_3 = binom.cdf(3, component, p_def)
p_at_least_3 = 1 - binom.cdf(3, component, p_def)
p_exactly_3, p_at_most_3, p_at_least_3
```

# 5. Objective: Demonstrate a complete data workflow: ETL to Visualization.
'''Dataset: OnlineRetail.csv'''

```python
Retail = pd.read_csv('OnlineRetail.csv',encoding='latin1')
Retail.head()
```

## 1. Data Cleaning & Preparation

```python
Retail.info()
```

```python
Retail_valid = Retail.dropna(subset=['CustomerID'])
Retail_valid = Retail_valid[Retail_valid['Quantity'] > 0]
```

```python
Retail_valid.info()
```

```python
Retail_valid['Revenue'] = Retail_valid['Quantity'] * Retail_valid['UnitPrice']
Retail_valid
```

## 2. Buisness Insights

```python
Retail_valid['InvoiceDate'] = pd.to_datetime(Retail_valid['InvoiceDate'])
Retail_valid['Month'] = Retail_valid['InvoiceDate'].dt.month
Retail_valid['Hour'] = Retail_valid['InvoiceDate'].dt.hour
Retail_valid
```

```python
rev = Retail_valid[Retail_valid['Country'] != 'United Kingdom'].groupby('Country')['Revenue'].sum()
rev_country = rev.idxmax()
rev_value = rev.max()
rev_country, rev_value
```

```python
top_products = Retail_valid.groupby('Description')['Revenue'].sum().sort_values(ascending=False).head(5)
top_products
```

```python
t_hour = Retail_valid.groupby('Hour')['InvoiceNo'].count()
peak_hour = t_hour.idxmax()
peak_hour
```

## 3. Exporting

```python
Retail_valid.to_csv('Processed_Sales_Data.csv')
```
