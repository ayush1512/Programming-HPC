```python
# Multiprocessing
from multiprocessing import Process, Pool

def square(n): 
    print(n**2)
    
def cube(n): 
    print(n**3)

p1 = Process(target=square, args=(4, ))
p2 = Process(target=cube, args=(4, ))

p1.start()
p2.start()

print("+- "* 25)
p1.join()
p2.join()
    
```

```python
import torch
import time

# Define a simple operation (e.g., matrix multiplication)
def perform_computation(device):
    a = torch.randn(10000, 10000, device=device)
    b = torch.randn(10000, 10000, device=device)
    start_time = time.time()
    c = torch.matmul(a, b)
    end_time = time.time()
    return end_time - start_time

# Test on CPU
cpu_time = perform_computation(torch.device("cpu"))
print(f"CPU computation time: {cpu_time:.4f} seconds")

# Test on MPS (if available)
if torch.backends.mps.is_available():
    mps_device = torch.device("mps")
    mps_time = perform_computation(mps_device)
    print(f"MPS computation time: {mps_time:.4f} seconds")
else:
    print("MPS not available for performance comparison.")
```
