```python

myFibDB = {}

def fib(n):
    global myFibDB

    if n in myFibDB:
        return myFibDB[n]
        
    if n == 0 :
        myFibDB[n] = 0
    elif n == 1:
        myFibDB[n] = 1
    else :
        myFibDB[n] = fib(n-1) + fib(n-2)
    return myFibDB[n]
        
#### App ###
def main():
    
    for val in range(36):
        print(f' fib({val}) = { fib(val)}', end=", ")
        
if __name__ == '__main__':
    import cProfile
    cProfile.run('main()')
    
```

### Co-routine

```python

def CountDown(Count):
    step = 1
    while Count >= 0:
        yield Count
        Count -= step
        
############ App ###
def main():
    for i in CountDown(10):
        print(f'doJob({i})')
        
if __name__ == '__main__':
    main()
```

```python

def CountDown(Count):
    step = 1
    while Count >= 0:
        step = yield Count
        Count -= step
        
############ App ###
def main():
    counter = CountDown(500)
    print(f'doJob({next(counter)})') # 500
    print("+- "*25)
    print(f'doJob({counter.send(75)})') # 425
    print("+- "*25)
    print(f'doJob({counter.send(150)})') # 425 - 150 => 275
    print("+- "*25)
    print(f'doJob({counter.send(175)})') # 100
    print("+- "*25)
    print(f'doJob({counter.send(100)})') #0
    print("+- "*25)
    try:
        print(f'doJob({counter.send(1)}') # exception
        print("+- "*25)
    except:
        print(' Counter completed ')
if __name__ == '__main__':
    main()
```

## Thread creation as object

```python

# Main-thread
def create_thread_from_class():
    class MyThread(threading.Thread):
        def __init__(self):
            super().__init__()

        def run(self):
            print(f'This thread-is created as a class-method')
            print(' Thread-ID = ', threading.get_native_id())
    
    #import threading

    print(' Main ID = ', threading.get_native_id())
    print(' Main Thread: os.getpid() = ', os.getpid())
    print(' Main Thread: os.getppid() = ', os.getppid())
    
    t1 = MyThread()

    t1.start()

    print(" Main waiting for child thread to complete")

    t1.join()
    
if __name__ == '__main__':
    create_thread_from_class()
```

```python

import os

# Child-Thread functions:
def thread_1(arg):
    print(' thread_1 start...: arg = ', arg)
    print(' Thread ID = ', threading.get_native_id())
    time.sleep(1)
    
    print(threading.get_native_id(), ' : thread_1 end...')
    
# Main-thread
def main():
    #import threading

    print(' Main ID = ', threading.get_native_id())
    
    t1 = threading.Thread(target = thread_1, args=( (10, 20, 30, 40), ) )
    t2 = threading.Thread(target = thread_1, args=( ([10, 20, 30], ) ))
    t3 = threading.Thread(target = thread_1, args=( ({10, 20}, ) ))
    t4 = threading.Thread(target = thread_1, args=( ({10:"TEN", 20:"TWENTY"}), ) )

    t1.start()
    t2.start()
    t3.start()
    t4.start()

    print(" Main waiting for child thread to complete")
    
    t1.join()
    t2.join()
    t3.join()
    t4.join()
    
if __name__ == '__main__':
    main()
```

```python

import os
import threading 
import time

# Child-Thread functions:
def thread_daemon(arg):
    print(' thread_1 start...: arg = ', arg)
    print(' Thread ID = ', threading.get_native_id())
    Count = 0
    while True:
        time.sleep(1)
        print(" This print message never gets printed ")
        Count += 1
        if Count % 47 == 0:
            break
        
# Main-thread
def main():

    print(' Main ID = ', threading.get_native_id())
    
    t1 = threading.Thread(target = thread_daemon, args=( (10, 20, 30, 40), ) )
    t1.daemon = True
    
    t1.start()
    print(" Main waiting for child thread to complete")
    
    #t1.join() # Once daemon-thread, it's always daemon
    
if __name__ == '__main__':
    main()
```

```python

def OtherProcess_Start():
    print(" Other Process start ---")
    print(' Main Thread: os.getpid() = ', os.getpid())
    print(' Main Thread: os.getppid() = ', os.getppid())
    print(" Other Process end ---")

# Main-Process
def main():
    #import threading

    print("Parent process start")
    print(' Main process ID = ', threading.get_native_id())
    print(' Main Thread: os.getpid() = ', os.getpid())
    print(' Main Thread: os.getppid() = ', os.getppid())
    
    p1 = multiprocessing.Process(target = OtherProcess_Start)

    p1.start()

    print(" Parent process waiting for child-process")
    p1.join()

    print("Parent process end")
    
if __name__ == '__main__':
    main()
```
